```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel(r'C:\Users\hp\Downloads\BPM.xlsx', sheet_name='Sheet1')
df

# Define the list of column pairs to plot in each subplot
column_pairs = [("IShimmer", "ISmartwatch"), 
                ("SShimmer", "SSmartwatch"),  
                ("StShimmer", "StSmartwatch"),  
               ("WShimmer", "WSmartwatch")]

# Create a 2 by 2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Loop over each subplot and plot the corresponding column pair
for i, ax in enumerate(axs.flat):
    if i < len(column_pairs):
        column1, column2 = column_pairs[i]
        df.plot(x="user", y=[column1, column2], ax=ax,marker="o",legend=False, color=['#1f77b4', '#ff7f0e'])

axs[0, 0].set_title("Ideal")        
axs[0, 1].set_title("Sitting")
axs[1, 0].set_title("Standing")
axs[1, 1].set_title("Walking")

# Create a legend using proxy artists
legend_colors = ['#1f77b4', '#ff7f0e']
legend_labels = ['Shimmer', 'Smartwatch']
handles = [plt.Line2D([], [], color=c) for c in legend_colors]
plt.legend(handles, legend_labels, loc='center left', bbox_to_anchor=(0.7,-0.25))


# Adjust the layout and show the plot
plt.tight_layout()
plt.savefig('heartcomp.png', dpi=800)
plt.show()
```
